```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
univ = pd.read_csv('/content/PCA.csv')
```

```python
univ
```

```python
univ1 = univ.iloc[:,1:]
```

```python
univ1
```

```python
uniarr = univ1.values
```

```python
uniarr
```

```python
uni_norm = scale(uniarr)
```

```python
uni_norm
```

```python
pca = PCA(n_components=6)
```

```python
pca_values= pca.fit_transform(uni_norm)
```

```python
pca_values
```

```python
var = pca.explained_variance_ratio_
```

```python
var
```

```python
varcum = np.cumsum(np.round(var,decimals=4)*100)
```

```python
varcum
```

```python
plt.plot(varcum,color='red')
```
