```python
books['User_ID'].value_counts()
```

```python
book_df = books.groupby('User_ID').count()['book_title']>=30
```

```python
book_df
```

```python
top_users = book_df[book_df].index
```

```python
top_users
```

```python
filtered_df = books[books['User_ID'].isin(top_users)]
```

```python
filtered_df
```

```python
a = filtered_df.pivot_table(index='book_title',columns='User_ID',values='book_rating')
```

```python
a
```

```python
pt = a
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi_scores = cosine_similarity(pt)
```

```python
simi_scores
```

```python
simi_scores.shape
```

```python
simi_scores[4]
```

```python
np.where(pt.index=='10 Commandments Of Dating')[0][0]
```

```python
from numpy.ma.core import sort
def recommend(book_name):
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(simi_scores[index])), key=lambda x:x[1],reverse=True)[1:6]
  for i in similar_items:
    print(pt.index[i[0]])


  
```

```python
recommend('10 Commandments Of Dating')
```

```python
recommend('1001 Brilliant Ways to Checkmate')
```
